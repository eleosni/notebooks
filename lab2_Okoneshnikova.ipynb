```python
swiss = pd.read_csv("https://gist.githubusercontent.com/christophsax/178d34245afdd6e187b1fff72dbe7448/raw/f5f4189f949f117bee4e82e4aa75c104ed20b4f4/swiss.csv")
```

```python
swiss.head()
```

```python
sns.set_theme(style="darkgrid")

g = sns.jointplot(x="Education", y="Infant.Mortality", data=swiss,
                  kind="reg", truncate=False,
                  xlim=(0, 60), ylim=(0, 20),
                  color="m", height=7)
```

```python

sns.set_theme(style="ticks")

sns.pairplot(swiss, hue="Location")
```

```python
import matplotlib.pyplot as plt
swiss.hist(edgecolor='black', linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(25,18)
plt.show
```

```python
sns.stripplot(x="Education", y="Infant.Mortality", data=swiss)
  
plt.show()
```

```python
sns.stripplot(x="Examination", y="Infant.Mortality", data=swiss)
  
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Infant.Mortality', data=swiss,orient='h')
plt.show()
```
